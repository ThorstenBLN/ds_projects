```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data from excel into Dataframe
lifeexp = pd.read_excel('../data/gapminder_lifeexpectancy.xlsx', index_col=0)
```

```python
# list of columns as columns selection
lifeexp = lifeexp[[1950, 1975, 2000, 2015]]
```

```python
# plot histplot in default mode and save the figure
lifeexp[2015].hist()
plt.savefig('histo.png')
```

```python
# plot histplot with 5 bins
lifeexp[2015].hist(bins=5)
```

```python
# plot histplot with 20 bins
lifeexp[2015].hist(bins=20)
```

```python
# plot customized histplot
plt.title('lifeexpectancy')
plt.xlabel('lifeexpectancy in years')
plt.ylabel('frequency')
plt.axis([45.0, 90.0, 0.0, 25.0])
lifeexp[2015].hist(bins=20)
```

```python
# plot customized histplot and save it as files
lifeexp[2015].hist(
    bins=20,
    facecolor='#ff0000',
    alpha=0.75,
    histtype='bar')
plt.savefig('histo.png', dpi=150)
plt.savefig('histo.svg', dpi=150)
```
